```python
renomear_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df.rename(columns=renomear_colunas, inplace=True)
df.columns
```

```python
df['senioridade'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
mapeamento_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(mapeamento_tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}
df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

```python
import numpy as np

#Criação da uma base df de teste
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000],
})

#Substitui os nulos pela media arredondada
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#Substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python

df_temperaturas = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

#O ffill preenche as colunas vazias com o valor da coluna anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

#O bfill preenche as colunas vazias com o valor da coluna posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'Nome': ['Ana','Bruno','Carlos','Daniele','Val'],
    'Cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém'],
})

#O fillna troca os campos 'vazios' pelo conteúdo inserido dentro do fillna
df_cidades['Cidade_preenchida'] = df_cidades['Cidade'].fillna('Não informado')

df_cidades
```

```python
#O dropna é utilizado para remover colunas com dados nulos
df_limpo = df.dropna()
```

```python
#Mostrando que o novo df (df_limpo) não contem nenhuma coluna com lacunas em branco (já que essas foram excluidas com o dropna)
df_limpo.isnull().sum()
```

```python
df_limpo.head()
```

```python
#O float64 considera númeors decimais e o int64 considera números inteiros. object são letras
df_limpo.info()
```

```python
df_limpo = df_limpo.assign(ano = df_limpo['ano'].astype('int64'))
```

# Aula 3 - Visualização de Dados

```python
df_limpo.head()
```

```python
df_limpo['senioridade'].value_counts().plot(kind = 'bar', title = 'Distribuição de Senioridade')
```

```python
import seaborn as sns
```

```python
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,6))
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
plt.title('Salário médio anual por nível de senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (usd)')
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().round(3).sort_values(ascending = False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().round(3).sort_values(ascending = False).index
ordem
```

```python
plt.figure(figsize = (8,6))
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd', order = ordem)
plt.title('Salário médio anual por nível de senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (usd)')
plt.show()
```

```python

plt.figure(figsize=(8, 4))
sns.histplot(df_limpo['usd'], bins=50, kde=True)
plt.title('Distribuição dos Salários')
plt.xlabel('Salário em USD')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = df_limpo['usd'])
plt.title('Boxplot salário')
plt.xlabel('Salário em usd')
plt.show()
```

```python
ordem_senioridade = ['Júnior', 'Pleno', 'Sênior', 'Executivo']

plt.figure(figsize = (8,5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade)
plt.title('Boxplot da distribuição por senioridade')
plt.xlabel('Salário em usd')
plt.show()
```

```python
ordem_senioridade = ['Júnior', 'Pleno', 'Sênior', 'Executivo']

plt.figure(figsize = (8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('Boxplot da distribuição por senioridade')
plt.xlabel('Salário em usd')
plt.show()
```

```python
import plotly.express as px
```

```python
df_senioridade_usd = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

fig = px.bar(
    df_senioridade_usd,
    x='senioridade',
    y='usd',
    title='Média Salarial por Nível de Senioridade',
    labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial (USD)'},
    category_orders={'senioridade': ordem_senioridade} # Usando a ordem já definida
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(
    remoto_contagem,
    names='tipo_trabalho',
    values='quantidade',
    title='Proporção dos tipos de trabalho'
)

fig.show()
```

```python
fig = px.pie(
    remoto_contagem,
    names='tipo_trabalho',
    values='quantidade',
    title='Proporção dos tipos de trabalho',
    hole=0.4
)

fig.update_traces(textinfo='percent+label')
fig.show()
```
