# Aula 1


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
new_columns = {
    'work_year': 'ano',
    'experience_level': 'experiencia',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'salario_usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localização',
    'company_size': 'tamanho'
}

df = df.rename(columns=new_columns)
df.columns
```

```python
experience_mapping = {
    'Senior': 'Sênior',
    'Mid-level': 'Pleno',
    'Entry-level': 'Júnior',
    'Executive': 'Executivo'
}
df['experiencia'] = df['experiencia'].replace(experience_mapping)
df.value_counts('experiencia')
```

```python
df.describe()
```

```python
df_remoto = df[df['taxa_remoto'] == 100]
```

```python
print(f"Número de profissionais que trabalham remotamente: {len(df_remoto)}")
```

import pandas as pd

# Supondo que você já tenha um DataFrame chamado df_salarios
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele'],
    'salario': [4000, None, 5000, None]
})

# Criando a coluna de status
def definir_status(salario):
    if salario is None:
        return 'Não informado'
    elif salario < 3000:
        return 'Baixo'
    elif 3000 <= salario < 7000:
        return 'Médio'
    else:
        return 'Alto'

df_salarios['status'] = df_salarios['salario'].apply(definir_status)

print(df_salarios)
